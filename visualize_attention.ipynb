### Visualize attention

```python
import numpy as np
import json
import matplotlib.pyplot as plt

## Load up the attention activation from the baseline model and the new model
old_att = np.load('terry_old_att.npy')
new_att = np.load('test_terry_new_att.npy')
context = np.load('context_terry_new.npy')
```

```python
context_ids = []
for (x) in context:
    for y in x:
        context_ids.append(y)
import json
with open("idx2word.json") as f:
    data = json.load(f)
words = []
count = 0
for val in context_ids:
    count += 1
#     if val == 5733:
#         print ("+++++++", data[str(val)], count)
#     if val == 9274:
#         print ("+++++++", data[str(val)], count)
    if str(val) in data:
        words.append(data[str(val)])
    else:
        words.append("unk")
```
